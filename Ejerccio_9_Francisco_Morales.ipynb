Modelo de Regresion Lineal

```python
import pandas as pd
```

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
datos= pd.read_excel("eje9.xlsx")
```

```python
datos.head
```

```python
from sklearn import linear_model
```

```python
modelo= linear_model.LinearRegression().fit(datos[["X"]],datos[["Y"]])
```

```python
modelo.intercept_, modelo.coef_
```

```python
mean_x = datos['X'].mean()
mean_x
mean_y = datos['Y'].mean()
mean_y

datos['X-Xbarra'] = datos['X']-mean_x
datos['Y-Ybarra'] = datos['Y']-mean_y
datos
#sum (xi-xbarra)(yi-yabrra)/n
datos['(X-Xbarra)(Y-Ybarra)'] = datos['X-Xbarra']*datos['Y-Ybarra']

covarianza = datos['(X-Xbarra)(Y-Ybarra)'].sum()/len(datos)
covarianza

sdx =np.std(datos['X'])
sdy =np.std(datos['Y'])

covarianza
```

```python
datos['Y_hat'] = modelo.predict(datos[['X']])
plt.scatter(datos[['X']],datos[['Y']])
plt.plot(datos[['X']],datos[['Y_hat']],color = 'r')
```

```python
datos['Error'] = datos['Y']-datos['Y_hat']
plt.hist(datos['Error'])
```

```python
from sklearn import metrics
```

```python
R2 = metrics.r2_score(datos[['Y']],datos[['Y_hat']])
R2
```

```python
ECM = metrics.mean_squared_error(datos[['Y']],datos[['Y_hat']])
ECM
```

```python
EMA = metrics.mean_absolute_error(datos[['Y']],datos[['Y_hat']])
EMA
```
